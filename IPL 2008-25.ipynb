<h1 style="color:red">Data Cleaning</h1>

```python
ipl.isnull().sum().sort_values(ascending=False).head(20)
```

```python
ipl.drop(columns=['team_reviewed', 'umpire', 'match_type', 'event_name','gender', 'team_type', 'balls_per_over', 'overs', 'event_match_no', 'match_number'], inplace=True)
```

```python
print(ipl.shape)
print(ipl.columns)
```

```python
ipl.isnull().sum().sort_values(ascending=False).head(20)
```

```python
#dealing with missing values and cleaning data
ipl['wicket_kind'] = ipl['wicket_kind'].fillna("None")
ipl['player_out'] = ipl['player_out'].fillna("None")
ipl['extra_type'] = ipl['extra_type'].fillna("None")

rare_cols = ['review_batter', 'review_decision', 'method', 'superover_winner']
for col in rare_cols:
    if col in ipl.columns:
        ipl[col] = ipl[col].fillna("Not Applicable")

ipl['runs_target'] = ipl['runs_target'].fillna(0)
ipl['win_outcome'] = ipl['win_outcome'].fillna("Unknown")

other_cols = ['fielders', 'new_batter', 'next_batter']
for col in other_cols:
    if col in ipl.columns:
        ipl[col] = ipl[col].fillna("None")

ipl['runs_total'] = pd.to_numeric(ipl['runs_total'], errors='coerce')
ipl['over'] = pd.to_numeric(ipl['over'], errors='coerce')


# FINAL CHECK
print("Remaining missing values:\n")
print(ipl.isnull().sum().sort_values(ascending=False).head(10))
```

```python
ipl['result_type'] = ipl.groupby('match_id')['result_type'].ffill()
ipl['result_type'] = ipl['result_type'].fillna("Unknown")

# FINAL CHECK
print("Remaining missing values:\n")
print(ipl.isnull().sum().sort_values(ascending=False).head(10))
```

<h3 style="color:brown">Creating Phase Column for PhaseWise Analysis</h3>

```python
def phase(over):
    if over <= 6:
        return "Powerplay Overs"
    elif over <= 15:
        return "Middle Overs"
    else:
        return "Death Overs"

ipl['phase'] = ipl['over'].apply(phase)
```

<h3 style="color:brown">PhaseWise Total Runs</h3>

```python
phase_runs = ipl.groupby('phase')['runs_total'].sum()
print(phase_runs)
```

<h3 style="color:brown">PhaseWise Average Run Rate</h3>

```python
phase_rr=(ipl.groupby('phase')['runs_total'].sum())/(ipl.groupby('phase')['phase'].count()/6)

print(phase_rr)
```

<h3 style="color:brown">PhaseWise Total Wickets Taken</h3>

```python
phase_wickets = ipl.groupby('phase')['team_wicket'].sum()

print(phase_wickets)
```

<h3 style="color:brown">PhaseWise Average Wicket Rate</h3>

```python
wicket_rate = ipl.groupby('phase')['team_wicket'].sum() / ipl.groupby('phase')['phase'].count()

print(wicket_rate)
```

<h1 style="color:red">Toss Analysis</h1>

<h3 style="color:brown">Creating A New Column 'toss_equals_winner'</h3>

```python
ipl['toss_equals_winner'] = (ipl['toss_winner'] == ipl['match_won_by'])
```

<h3 style="color:brown">Checking Impact of Winning Toss</h3>

```python
print("Success Percent By Toss decison:")
win_percent_by_toss=((ipl[ipl['toss_equals_winner']==True]['toss_equals_winner'].count())/ipl.shape[0])*100
print(win_percent_by_toss,"\n")

print("Failure Percent By Toss decison:")
print(100-win_percent_by_toss)

```

<h5 style="color:blue">From the analysis conducted above, we can see that the toss does't have a significant impact on the outcome of the match, it relies entirely on <b><a style="color:black">skill and strategy</a></b>.<br>The chances of winning the game are <b><a style="color:black">approximately 50-50</a></b>, though the probability of winning tilts <b><a style="color:black">slightly in favor of the team who wins the toss</a><b>.  </h5>
